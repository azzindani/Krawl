```python
# Cell 2 — Run the engine then query results
import subprocess, json, os, time

# ── Run ───────────────────────────────────────────────────────────
print("="*60)
print("RUNNING KRAWL — 100 TARGETS")
print("="*60)
print("Watch live progress below...\n")

t0 = time.time()

subprocess.run(
    ["bash", "run.sh", "tasks/test_100.json"],
    check=False
)

elapsed = time.time() - t0
print(f"\nTotal time: {elapsed:.0f}s ({elapsed/60:.1f} min)")

# ── Stats ─────────────────────────────────────────────────────────
print("\n" + "="*60)
print("DATABASE STATS")
print("="*60)
subprocess.run(["npx","tsx","krawl.ts","--stats","--db","krawl.db"])

# ── Key queries ───────────────────────────────────────────────────
QUERIES = {
    "Results by mode": """
        SELECT mode,
               COUNT(*) as total,
               SUM(CASE WHEN status='ok' THEN 1 ELSE 0 END) as ok,
               SUM(CASE WHEN status!='ok' THEN 1 ELSE 0 END) as failed,
               ROUND(AVG(elapsed_ms),0) as avg_ms
        FROM pages GROUP BY mode ORDER BY total DESC
    """,
    "Stock prices (latest)": """
        SELECT ticker, price, change_val, change_pct, currency
        FROM stocks ORDER BY extracted_at DESC LIMIT 15
    """,
    "News headlines": """
        SELECT source, headline FROM news
        ORDER BY extracted_at DESC LIMIT 10
    """,
    "Market indices": """
        SELECT index_name, price, change_val, change_pct
        FROM market_indices ORDER BY extracted_at DESC
    """,
    "Files discovered": """
        SELECT ext, COUNT(*) as count
        FROM files GROUP BY ext ORDER BY count DESC
    """,
    "Domain routing cache": """
        SELECT domain, mode, total_pages, total_errors,
               ROUND(avg_ms,0) as avg_ms
        FROM domains ORDER BY total_pages DESC LIMIT 15
    """,
}

for label, sql in QUERIES.items():
    print(f"\n{'─'*60}")
    print(f"{label}")
    print("─"*60)
    r = subprocess.run(
        ["npx","tsx","krawl.ts","--query",sql.strip(),"--db","krawl.db"],
        capture_output=True, text=True
    )
    try:
        rows = json.loads(r.stdout)
        if not rows:
            print("  (empty)")
            continue
        cols   = list(rows[0].keys())
        widths = {c: min(max(len(c), max(len(str(row.get(c,""))) for row in rows)), 30) for c in cols}
        header = "  ".join(str(c).ljust(widths[c]) for c in cols)
        print(header)
        print("-" * len(header))
        for row in rows:
            print("  ".join(str(row.get(c,"")).ljust(widths[c])[:widths[c]] for c in cols))
        print(f"  ({len(rows)} rows)")
    except Exception:
        print(r.stdout[:400] or r.stderr[:200])

# ── Full-text search ──────────────────────────────────────────────
print(f"\n{'─'*60}")
print("FULL-TEXT SEARCH DEMOS")
print("─"*60)

for table, query in [("fts_news","saham"),("fts_news","market"),("fts_pages","bank")]:
    r = subprocess.run(
        ["npx","tsx","krawl.ts","--search",query,"--db","krawl.db"],
        capture_output=True, text=True
    )
    lines = [l for l in r.stdout.split("\n") if l.strip()]
    print(f"\n  search '{query}' → {len([l for l in lines if 'URL' in l or 'headline' in l.lower()])} results")
    for line in lines[:8]:
        print(f"  {line}")

# ── Export ────────────────────────────────────────────────────────
print(f"\n{'─'*60}")
print("EXPORTING TO CSV")
print("─"*60)
subprocess.run(
    ["npx","tsx","krawl.ts","--export","all","--out","exports","--db","krawl.db"]
)

print(f"\nExported files:")
if os.path.exists("exports"):
    for f in sorted(os.listdir("exports")):
        size = os.path.getsize(f"exports/{f}")
        print(f"  {f:<30} {size:>10,} bytes")

# ── Output files ──────────────────────────────────────────────────
print(f"\n{'─'*60}")
print("OUTPUT FILES")
print("─"*60)
for fname in ["krawl.db","krawl_output.jsonl","krawl_checkpoint_default.json","router_cache.json"]:
    if os.path.exists(fname):
        size = os.path.getsize(fname)
        print(f"  ✓ {fname:<45} {size:>10,} bytes")
    else:
        print(f"  ✗ {fname}")

if os.path.exists("krawl_output.jsonl"):
    with open("krawl_output.jsonl") as f:
        lines = sum(1 for l in f if l.strip())
    print(f"\n  JSONL records: {lines:,}")
```
